```python
#importing packages
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
data.head()
```

```python
df=data.drop(['User ID'],axis=1) #axis=1 will delete the entire column
```

```python
df.head()
```

```python
df['Gender']=df['Gender'].apply({'Male':1,'Female':2}.get)
```

```python
df.head()
```

```python
x=df[['Gender','Age','EstimatedSalary']] #independent
y=df['Purchased'] #dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7) #K=5
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
#checking the accuracy with the help of ConfusionMatrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
ac=accuracy_score(y_test,y_pred)
```

```python
print(cm)
```

```python
print(ac)
```
